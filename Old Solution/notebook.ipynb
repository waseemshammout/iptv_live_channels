```python
import requests
url = "https://api.github.com/repos/iptv-org/iptv/contents/streams"
response = requests.get(url)
if response.status_code == 200:
    files = response.json()
    for file_info in files:
        file_url = file_info["download_url"]
        file_response = requests.get(file_url)
        if file_response.status_code == 200:
            for line in file_response.text.splitlines():
                if line.find('#EXTM3U') == -1 and line.find('EXTVLCOPT') == -1:
                    line = line.strip()
                    with open('t.txt', 'a', encoding='utf-8') as txtFile:
                        txtFile.write(line + '\n')
                        txtFile.close()
else:
    print("Failed to retrieve file list:", response.status_code)
```

```python
import csv
input_file = "t.txt"
output_file = "channels.csv"
rows = []
with open(input_file, "r", encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip()]
i = 0
while i < len(lines):
    line = lines[i]
    if line.startswith("#EXTINF"):
        channel_name = line.split(",", 1)[1]
        if i + 1 < len(lines):
            url = lines[i + 1]
            rows.append([channel_name, url])
            i += 2
        else:
            i += 1
    else:
        i += 1
with open(output_file, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Channel Name", "URL"])  # header
    writer.writerows(rows)
print(f"CSV file created: {output_file}")
```
